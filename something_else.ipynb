```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('df_factors.csv')

data.head()
```

```python
#remove not genetic counts
#data = data.drop(columns=['gender', 'country', 'occupation'])

# Group by all columns and count the occurrences
counts = data.groupby(list(data.columns)).size().reset_index(name='count')

sorted_counts = counts.sort_values(by='count', ascending=False)

# Calculate the total number of rows
total_rows = len(data)

# Calculate the percentage for each combination
counts['percentage'] = (counts['count'] / total_rows) * 100

counts.head()
sorted_counts = counts.sort_values(by='percentage', ascending=False)

# Display the result
print(counts)

sorted_counts.to_csv('count_factor.csv', index=False)
```

```python
#remove not genetic counts
data_main = pd.read_csv('clean_df_main.csv')

# Group by all columns and count the occurrences
counted = data_main.groupby(list(data_main.columns)).size().reset_index(name='count')

sorted_counts = counted.sort_values(by='count', ascending=False)

# Display the result
print(counted)

sorted_counts.to_csv('df_init_count.csv', index=False)
```

```python
#discount gender, country and occupation

discount_1 = data_main.drop(columns=['gender', 'country', 'occupation', 'self_employed'])

# Group by all columns and count the occurrences
counted = discount_1.groupby(list(discount_1.columns)).size().reset_index(name='count')

sorted_counts = counted.sort_values(by='count', ascending=False)

# Display the result
print(counted)

sorted_counts.to_csv('df_init_count.csv', index=False)
```
