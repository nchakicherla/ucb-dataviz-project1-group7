```python
# Load data from CSV into a DataFrame
df_factors = pd.read_csv('df_factors.csv')

# Calculate a unique candidacy DataFrame from severity 
df_candidacy = pd.DataFrame()
df_candidacy['candidacy'] = (df_factors['family_history'] +
                             df_factors['treatment'] +
                             df_factors['growing_stress'] +
                             df_factors['changes_habits'] +
                             df_factors['mental_health_history'])

# Calculate unique severity DataFrame from candicdicay
df_severity = pd.DataFrame()
df_severity['severity'] = (df_factors['growing_stress'] +
                           df_factors['changes_habits'] +
                           df_factors['coping_struggles'] +
                           df_factors['work_interest'] +
                           df_factors['social_weakness'])

# Save these DataFrames to CSV files
df_candidacy.to_csv('df_candidacy.csv', index=False)
df_severity.to_csv('df_severity.csv', index=False)

# Display first few rows of candidacy and severity DataFrames
print("Candidacy DataFrame:")
print(df_candidacy.head())

print("\Severity DataFrame:")
print(df_severity.head())
```

```python
# Display first few rows of candidacy DataFrame
print("Candidacy DataFrame:")
print(df_candidacy.head())

# Display first few rows of severity DataFrame
print("\nSeverity DataFrame:")
print(df_severity.head())

# Aggregate by MH Candidacy to discern frequency of unique values/size
mh_candidacy_summary = df_candidacy.groupby('candidacy').size()

# Aggregate by MH Severity
mh_severity_summary = df_severity.groupby('severity').size()

# Print summary statistics or perform further analysis
print("\nMH Candidacy Summary:")
print(mh_candidacy_summary)

print("\nMH Severity Summary:")
print(mh_severity_summary)
```
