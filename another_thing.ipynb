```python
import pandas as pd
import itertools

# Load your data into a DataFrame
# Replace 'your_data.csv' with your actual file path
data = pd.read_csv('clean_df_main.csv')

def calculate_commonness(df):
    # Group by all columns and count the occurrences
    counts = df.groupby(list(df.columns)).size().reset_index(name='count')
    total_rows = len(df)
    # Calculate the percentage for each combination
    counts['percentage'] = (counts['count'] / total_rows) * 100
    return counts

def find_commonness_for_permutation(data, columns, target_percentage=5.0):
    current_columns = list(columns)
    removed_columns = []
    
    while current_columns:
        # Calculate commonness with the current set of columns
        filtered_data = data[current_columns]
        commonness = calculate_commonness(filtered_data)
        
        # Check if any combination meets or exceeds the target percentage
        if commonness['percentage'].max() >= target_percentage:
            return commonness, removed_columns, True
        
        # If not, remove one column and continue
        removed_columns.append(current_columns.pop())
    
    return commonness, removed_columns, False

def remove_columns_until_commonness(data, target_percentage=10.0):
    columns = list(data.columns)
    column_permutations = itertools.permutations(columns)
    
    for perm in column_permutations:
        commonness, removed_columns, success = find_commonness_for_permutation(data, perm, target_percentage)
        if success:
            print(f"Reached target commonness with columns: {perm[:len(perm) - len(removed_columns)]}")
            return commonness, removed_columns

    print("Could not reach target commonness by removing columns in any order.")
    return calculate_commonness(data[columns]), columns

# Call the function and get the result
commonness, removed_columns = remove_columns_until_commonness(data)

# Display the result
print(commonness)
print(f"Removed columns: {removed_columns}")
```

```python
import pandas as pd
from itertools import combinations

# Load the data
df_main = pd.read_csv('df_factors.csv')
df_refresh = df_main.copy()

# Drop the specified columns
columns_to_consider = ['social_weakness', 'days_indoors']
#data_refresh = data_refresh.drop(columns=columns_to_consider)

# Calculate total rows for percentage calculation
total_rows = len(data_refresh)

# Function to find the best columns to drop
def find_best_columns_to_drop(data, columns_to_consider, max_columns_to_drop, min_percentage):
    best_percentage = 0
    best_columns_to_drop = []
    
    for i in range(1, max_columns_to_drop + 1):
        for combo in combinations(columns_to_consider, i):
            temp_data = data.drop(columns=list(combo))
            counted = temp_data.groupby(list(temp_data.columns)).size().reset_index(name='count')
            retained_percentage = (counted['count'].max() / total_rows) * 100
            if retained_percentage > best_percentage:
                best_percentage = retained_percentage
                best_columns_to_drop = list(combo)
            if retained_percentage > min_percentage:
                return best_columns_to_drop, retained_percentage
    
    return best_columns_to_drop, best_percentage

max_columns_to_drop = len(columns_to_consider)  # You can set this to a lower number if needed
min_percentage = 5  # Minimum percentage threshold

best_columns_to_drop, best_percentage = find_best_columns_to_drop(data_refresh, columns_to_consider, max_columns_to_drop, min_percentage)

if best_percentage > min_percentage:
    print(f"Best columns to drop: {best_columns_to_drop}")
    print(f"Best percentage retained: {best_percentage:.2f}%")
else:
    print("No combination of columns meets the minimum percentage threshold.")
```
